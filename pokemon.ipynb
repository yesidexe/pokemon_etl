```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_df = pd.read_csv('data/Pokemon.csv')
pokemon_df.head()
```

```python
pokemon_df.shape
```

```python
pokemon_df.isnull().sum()/len(pokemon_df)*100
```

```python
pokemon_df['type2'] = pokemon_df['type2'].fillna(pokemon_df['type1'])
```

```python
pokemon_df.duplicated().sum()
```

```python
pokemon_df['type2'].value_counts()
```

```python
# pokemon_df['type2'].value_counts().plot(kind='bar')
# plt.xticks(rotation=45)
# plt.show()
```

```python
pokemon_df['type1'] = pokemon_df['type1'].str.replace('Graass','Grass', regex=False)
pokemon_df['type2'] = pokemon_df['type2'].str.replace('Graass','Grass', regex=False)
```

```python
water_pokemon = pokemon_df.copy()
water_pokemon = pokemon_df[(pokemon_df['type1'] == 'Water') | (pokemon_df['type2'] == 'Water')]
water_pokemon.sample(10)
```

```python
ground_pokemon = pokemon_df.copy()
ground_pokemon = pokemon_df[(pokemon_df['type1'] == 'Ground') | (pokemon_df['type2'] == 'Ground')]
ground_pokemon.sample(10)
```

```python
print(len(water_pokemon))
print(len(ground_pokemon))
```

```python
pokemons = pd.concat([water_pokemon,ground_pokemon])
pokemons = pokemons.drop_duplicates(subset='name')
```

```python
pokemons['name'].value_counts()
```

```python
pokemons['is_strong'] = pokemons['attack'].apply(lambda x: 'Es fuerte' if x > 60 else 'Es debil')
pokemons.head()
```

```python
pokemons
```

```python
pokemons.to_csv('data/pokemons.csv',index=False)
```
